```
'''Use ODR to fit a linear function to data with errors in
   both x and y.'''
#This part is used for drawing the graphs and finding the errors

from scipy.odr import *

import random

# Initiate some data, giving some randomness using random.random().
y = kv45_lnminsi_9
x = kv45_Ti_9

x_err =  [math.sqrt(2)/2,math.sqrt(2)/2,math.sqrt(2)/2,math.sqrt(2)/2,math.sqrt(2)/2,math.sqrt(2)/2]

y_err = kv45_lnminsi_sigmas_9
# Define a function (quadratic in our case) to fit the data with.
def linear_func(p, x):
	 m, c = p
	 return m*x + c

# Create a model for fitting.
linear_model = Model(linear_func)

# Create a RealData object using our initiated data from above.
data = RealData(x, y, sy=y_err)

# Set up ODR with the model and data.
odr = ODR(data, linear_model, beta0=[0., 1.])

# Run the regression.
out = odr.run()

# Use the in-built pprint method to give us results.
out.pprint()

# Generate fitted data.
x_fit = np.linspace(x[0], x[-1], 1000)
y_fit = linear_func(out.beta, x_fit)


# Generate a plot to show the data, errors, and fit.
fig, ax = plt.subplots()

ax.errorbar(x, y, yerr=y_err, linestyle='None', marker='x')
ax.plot(x_fit, y_fit)

ax.set_xlabel(r'$lambdas(dimesnionless)$')
ax.set_ylabel(r'$Kilovolt values(dimensionless)$')
ax.set_title('11 squeze lambdas versus kilovolt values')

plt.show()
```

```
#These are error propagation formulas for different mathematical calculations. ı took them from wikipedia_error propagation
def multiplicity_err_prp(x,y,z,t):  #x is the first number y is its uncertainity, 
  f = x*z
  k = f*math.sqrt((y*y/x**2) + (t*t/z**2))

  return f,k

def subtraction_err_prp(x,y,z,t):
  f = x-z
  k = math.sqrt(y*y + z*z)
  return f,k

def division_err_prp(x,y,z,t):
  f = x/z
  k = f*math.sqrt((y*y/x**2) + (t*t/z**2))
  return f,k

def summation_err_prp(x,y,z,t):
  f = x + z
  k = math.sqrt(y*y + z*z)
  return f,k

def sine_err_prp(a,b,A,sigma):
  k = a * b * (np.cos(np.deg2rad(b*A)))*sigma
  return k

def ln_err_prp(a,A,sigma):
  k = abs(math.sqrt((a * sigma / A)*2))
  return k

```
